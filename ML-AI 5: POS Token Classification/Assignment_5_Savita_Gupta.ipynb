**Instructions:**: Figure out how to use pipelines() to do POS - Part of Speech token classification. Give a few examples.

*I used this documentation from huggingface for reference: https://huggingface.co/docs/transformers/main_classes/pipelines*

```python
from transformers import pipeline
```

```python
POStest = pipeline(model="vblagoje/bert-english-uncased-finetuned-pos", aggregation_strategy="simple")
```

**Examples:**

```python
POStest("Hi, my name is Savita and I am a student in ADSP 32017.")
```

```python
POStest("I've always dreamed of living in London.")
```

```python
POStest("I work full time for a startup called Pattern Health, and I am almost finished with the Biomedical Informatics masters program.")
```

```python
POStest("I have a brother named Tivas. His name is my name (Savita) spelled backwards without the 'a'.")
```
